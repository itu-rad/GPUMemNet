```python
# import random
import matplotlib.pyplot as plt
from numpy import random

depth_min = 1
depth_max = 20


batch_size_min = 2
batch_size_max = 1024

dlist = []
blist = []

random.seed(99)

for i in range(1, 2001):
    # Randomly generate depth, width, and batch size
    depth = random.uniform(depth_min, depth_max)
    batch_size = random.uniform(batch_size_min, batch_size_max)
    dlist.append(depth)
    blist.append(batch_size)

plt.hist(dlist, bins=100, edgecolor='black')

# Add titles and labels
plt.title('Histogram of Depth Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()


plt.hist(blist, bins=100, edgecolor='black')

# Add titles and labels
plt.title('Histogram of Batch_Size Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# import random
import matplotlib.pyplot as plt
import numpy as np

il = []
ol = []
bl = []
dl = []
al = []

for i in range(2000):
  input_size = int(np.random.uniform(4, 4096))  # Random input size between 32 and 4096
  output_size = int(np.random.uniform(1, max(2, input_size // 4)))  # Random output size, smaller than input size
  batch_size = int(np.random.uniform(1, 256) * 4)
  depth = int(np.random.uniform(1, 11))  # Random depth (number of hidden layers)
  architecture = np.random.choice(['pyramid', 'uniform', 'bottleneck', 'gradual'])  # Random architecture type

  il.append(input_size)
  ol.append(output_size)
  bl.append(batch_size)
  dl.append(depth)
  al.append(architecture)


plt.hist(il, bins=100, edgecolor='black')

# Add titles and labels
plt.title('Histogram of Input Size')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()


plt.hist(ol, bins=100, edgecolor='black')

# Add titles and labels
plt.title('Histogram of Output Size')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()


plt.hist(bl, bins=100, edgecolor='black')

# Add titles and labels
plt.title('Histogram of Batch Size')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()



plt.hist(dl, bins=100, edgecolor='black')

# Add titles and labels
plt.title('Histogram of Depth')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()

plt.hist(al, bins=100, edgecolor='black')

# Add titles and labels
plt.title('Histogram of Architecture')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
